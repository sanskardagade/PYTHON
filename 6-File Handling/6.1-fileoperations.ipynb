**File Operation - Read and Write Files**

File handeling is a crucial part of programming language. Python provides built-in functions and methods to read from and write to files, both text and binary. This lessons will cover the basics of the handling, including reading and writing text files and binary files.

```python
##Read a whole file 

with open('example.txt', 'r')as file:
    content = file.read()
    print(content)
```

```python
##Read a file line by line 
with open('example.txt','r')as file:
    for line in file:
        print(line)
```

```python
##Read a file line by line 
with open('example.txt','r')as file:    
    for line in file:
        print(line.strip())   ##strip() removes the newline character 
        
```

```python
##Writring in a file

with open('example.txt','w') as file:
    file.write("hello world\n")
    file.write("i am sanskar\n")
```

```python
##Write a file without Overwriting 

with open('example.txt','a') as file:
    file.write("Append operation taling place")
    
```

```python
##Writing a list of lines to a file

lines = ['\nFirst Line \n' , 'Second Line\n' , 'Third Line \n']
with open ('example.txt','a') as file:
    file.writelines(lines)
```

```python
###Binary Files

##Writing to a binary file

data = b'\x00\x01\x02\x03\x04'
with open('example.bin','wb') as file:
    file.write(data)
```

```python
##Reading a binary file

with open('example.bin','rb') as file:
    content = file.read()
    print(content)


## in our language
# with open('example.bin','rb') as file:
#     content = file.read()
#     text = content.decode('utf-8')   # or 'ascii'
#     print(text)
```

```python
##Read a content from source text file and write it to the destination text file

with open('example.txt','r')as source_file:
    content = source_file.read()

with open('destination.txt','w')as destination_file:
    destination_file.write(content)
```

```python
##Read a text file and count the number of lines, words and characters

def count_text_file(file_path):
    with open(file_path,'r')as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return line_count,word_count,char_count

file_path = 'example.txt'
lines, words, characters = count_text_file(file_path)
print(f'Lines:{lines}, Words:{words}, Characters:{characters}') 
```

```python
###Writing and then reading a file 
## w+ mode in python is used to open a file for both writing and reading.If the file does not exist, it creates a new file. If the file already exists, it truncates the file to zero length, effectively erasing its contents before writing.(i.e the file is overwritten)
with open('example.txt','w+') as file:
    file.write("Hello world.\n")
    file.write("This is a new line.\n")

    ##Move the file cursor to the beginning of the file before reading
    file.seek(0)

    ##Read the content of the file
    content = file.read()
    print(content)
```
