*Importing Modules and Packages in Python*

In python,modules and packagess help to organize and reuse code. Here's a comprehensive guide on how to import them.

### Python is open source

```python
import math
math.sqrt(16)
```

```python
from math import sqrt,pi
print(sqrt(25))
print(sqrt(100))
print(pi)
```

```python
import numpy as np
np.array([1,2,3])
```

```python
from math import * ## * indicates we are able to access all thr funcions inside the library math
sqrt(100)
```

```python
from package.maths import addition
addition(2,3)
```

```python
from package import maths
maths.addition(3,2)

```
