## Iterators

Iterators are advanced Python concepts that allow for efficeint looping and memory management. Iterators provide a way to access elements of a collection sequentially without exposing the underlying structure.

```python
my_list = [1,2,3,4,5,6]
for i in my_list:
    print(i)

    
```

```python
type(my_list)
```

```python
print(my_list)
```

```python
##Iterator 
iterator = iter(my_list)
print(type(iterator))
```

```python
iterator                ##it uses lazy loading 
```

```python
## Iterate through all the elements

next(iterator)
```

```python
iterator = iter(my_list)
```

```python

try:
    print(next(iterator))
except StopIteration:
    print("There are no elements in the iterator")
```

```python
##String iterator

my_string = "Hello"
string_iterator = iter(my_string)

print(next(string_iterator))
print(next(string_iterator))
```
