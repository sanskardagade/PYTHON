## Python Logging 


Logging is a crucial aspecct of any application, providing a way to track events, errors and operational information. Ppython's built-in logging module offers a felxible framework for emitting log messages from Python Programs. In this lesson , we will cover the basics of logging, including how to configure logging, log levels and best practices for using logging in python application. 

```python
import logging

## Condfigure the basic logging settings

logging.basicConfig(level=logging.DEBUG)

## log messages with different severity levels

logging.debug("This is a debug message")
logging.info("This is an info message")
logging.warning("This is a warning message")
logging.error("This is an error message")
logging.critical("This is a critical message")
```

### Log Levels

Pythons's logging module has several log levels indicating the severity of events. The default levels are:

###### DEBUG : Detailed information, typically of interest only when diagnosing problems. 
###### INFO : Confirmation that things are working as expected.
###### WARNING : An indication hat something unexpected happened or indicative of some problems in the near future (e.g., 'disk space low'). This software is still working as expected.
###### ERROR : Due to a more serious problem, the software has not been able to perform some function.
###### CRITICAL : A very serious error , indiacting that the program itself may be unable to continue reading .


```python
## configuring the logging 

logging.basicConfig(
    level = logging.DEBUG,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    datefmt='%Y-%m-%d %H:%M;%S',

    # force=True
)


logging.debug("This is a debug message")
logging.info("This is an info message")
logging.warning("This is a warning message")
logging.error("This is an error message")
logging.critical("This is a critical message")
```

```python
## configuring the logging 
import logging
logging.basicConfig(
    filename='app.log',
    filemode='w',
    level = logging.DEBUG,                  ##when you are changing the mode you need to restart the kernel 
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    datefmt='%Y-%m-%d %H:%M;%S',
    force=True
)


logging.debug("This is a debug message")
logging.info("This is an info message")
logging.warning("This is a warning message")
logging.error("This is an error message")
logging.critical("This is a critical message")
```
