## Logging with multiple loggers

You can create multiple loggers for different parts if your application .

```python
import logging

##create a logger for module1

logger1 = logging.getLogger("module1")
logger1.setLevel(logging.DEBUG)

##create a logger for module 2

logger2 = logging.getLogger("module2")
logger2.setLevel(logging.WARNING)

##Configure logging setting 

logging.basicConfig(
    level = logging.DEBUG,
    format='%(asctime)s - %(name)s - %(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)
```

```python
## log message with different loggers

logger1.debug("This is a debug message for module1")
logger2.warning("This is warning message for logger2")
logger2.error("This is the error message")
```
