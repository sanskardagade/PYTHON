***Tuples***

*Introduction to Tuples*

Tuples are ordered collections of items that are immuatable. They are similar to lists,but their immutability makes them different.

```python
##Creating a Tuple
empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst = list()
print(type(lst))
tpl = tuple()
print(type(tpl))
```

```python
numbers = tuple([1,2,3,4,5,6])
print(numbers)
print(type(numbers))
```

```python
da = list((1,2,3,4,5,6))
print(da)
print(type(da))
```

```python
mixed_tuple = (1,"Hello World",3.14,True)
print(mixed_tuple)
```

```python
##Accessing tuple elements
numbers[::-1]
numbers[0:4]
numbers[::1]

```

```python
##Tuples operations

numbers + mixed_tuple    ## concatination
```

```python
mixed_tuple * 3
```

```python
numbers * 3
```

```python
##Immutable nature of tuples
##Tuples are immutable, means their elements cannot be changed once assigned 

lst = [1,2,3,4,5]
print(lst)
lst[1] = "sanskar"
print(lst)
```

```python
##Tuples Methods

print(numbers.count(9))
print(numbers.index(2))
```

```python
##Packing and Unpacking Tuples
##packing a tuple
packed_tuple = 1,"Hello",3.14
print(packed_tuple)
```

```python
##Unpacking Tuple 
a,b,c = packed_tuple
print(a)
print(b)
print(c)
```

```python
##Unpacking with *

numbers = (1,2,3,4,5,6)
first,*middle,last = numbers
print(first)
print(middle)
print(last)
```

```python
##Nested Tuple
##Nested List

lst = [[1,2,3,4],[5,6,7],[8,9],[True,"sanskar"]]
print(lst[0])
print(lst[0][1])
print(lst[0][0:3])
```

```python
##Inside a list we can have a tuple

abc = [[1,2,3],(4,5,6),[7,"snd",8.33]]
print(abc[2])
print(abc[2][2])
```

```python
nested_tuple = ((1,2,3),("a","b","c"),(8,9))
print(nested_tuple[1])
print(nested_tuple[0][1])
```

```python
##Iterating over nested tuples

for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")
    print()
```

Tuples are versatile and useful in  many real-world scenarios where an immutable and ordered collection of items is required. They are commonly used in data structures, function and arguments and return values, and as dictionary keys. Understanding how to leverage tuples effectively can improve the efficiency and readability of your python code.
