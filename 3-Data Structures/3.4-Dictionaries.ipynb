***Dictionaries***

Dictionaries are unordered collection of items. They store data in key-value pairs. Keys must be unique and immutable(eg. Strings, numbers and tuples), while values can be of any type.


```python
dd = ()
print(type(dd))  # <class 'tuple'>
tt = []
print(type(tt))
ss = {}
print(type(ss))
empty_set = set()
print(type(empty_set))
```

```python
empty_dict = dict()
print(empty_dict)
print(type(empty_dict))
```

```python
student = {"name":"Sanskar", "age":20, "grade":24}  ##Dictionary is of key-value pair
print(student)
print(type(student))
```

```python
##Single key is always used
student = {"name":"sanskar", "age":32, "name":24}  ##The latest name gets replaced so key must be unique
print(student)
```

```python
##Accessing Dictionary Elements
student = {"name":"Sanskar","age":32,"grade":"A"}
print(student)
```

```python
##Accessing Dictionary Elements
print(student["name"])
print(student["age"])
print(student["grade"])
## print(student["nav"])                ##This gives error 


##Accessing using get() method

print(student.get("name"))
print(student.get("age"))
print(student.get("grade"))
print(student.get("nav"))           ##But if we use built in method then we get the result
print(student.get("nav","Not Available")) ##Default name to give
```

```python
##Modifying dictionary elements
##Dictionary are mutable, so can add, modify and delete element


print(student)
```

```python
print(student)                              ##Updated value for the key
student["age"] = 33
print(student)
student["address"] = "India"                ##Added a new key and value 
print(student)

```

```python
##For deleting a element we have a keyword del for it 

del student["grade"]                        ##Delete key and value pair
print(student)
```

```python
##Dictionary Methods

keys = student.keys()                                   ##Get all the keys
print(keys)
print(type(keys))
values = student.values()                               ##Get all values
print(values)   
print(type(values))
items = student.items()                                 ##Get all key-value pairs in the form of tuples
print(items)
print(type(items))
```

```python
##Shallow Copy 

student_copy = student 
print(student)
print(student_copy)

```

```python
student["name"] = "SSS"
print(student)
print(student_copy)
```

```python
student_copy1 = student.copy()              ##This is how we make shallow copy it provides different memory loaction
print(student_copy1)
print(student)
```

```python
student["name"] = "JJJJ"
print(student_copy1)
print(student)
```

```python
##Acessing over dictionaries
##You can use loops to iterate over dictionaries, key, values or items


##Iterating over keys
for keys in student.keys():
    print(keys)
```

```python
##Iterate over values 
for value in student.values():
    print(value)
```

```python
##Iterate over key value pairs 
for key,value in student.items():
    print(f"{key}:{value}")
```

```python
##Nested Dictionaries ( A dictionary inside another dictionary)
students={
    "student1":{"name":"Sans","age":32},
    "student2":{"name":"kar","age":40}
}
print(students)
print(type(students))
```

```python
##Access Nested dictionaries elements

print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
students.items()
```

```python

# Iterating over nested dictionaries
for student_id, student_info in students.items():
    print(f"{student_id}: {student_info}")
    for key, value in student_info.items():
        print(f"{key}: {value}")

        
```

```python
##Dictionary Comprehension

squares = {x:x**2 for x in range(5)}
print(squares)
```

```python
##Conditional Dictionary Comprehesnion
even = {x:x**2 for x in range(10) if x%2==0}
print(even)
```

```python
##Practical Examples 

##Use a dictionary to count the frequency of elements in list
numbers = [1,2,2,3,3,3,4,4,4,4]
frequency = {}

for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number] = 1

print(frequency)
```

```python
## Merge two dictionaries into one 
dict1 = {"a":1, "b":2}
dict2 = {"b":3, "d":4}
merged_dict = {**dict1,**dict2}
print(merged_dict)
```
