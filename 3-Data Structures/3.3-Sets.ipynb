****Sets****

Sets are built-in data types in Python used to store collection of unique items. They are unordered, meaning that the elements do not follow a specific order, and they do not allow duplicate elements. Sets are useful for membership tests, eliminating duplicate entries, and performing mathematical set operations like union,intersection, difference and symmetric difference.

```python
##Creating a Set
my_set = {1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
my_empty_set = set()
print(type(my_empty_set))
```

```python
##Typecast with the help of set keyword

my_set = set([1,2,3,4,5,6,7,7])
print(my_set)
```

```python
##Basic set operations
##Adding and removing elements
my_set.add(7)
print(my_set)
my_set.add(7)
print(my_set)
```

```python
##Remove the element from the set
my_set.remove(3)
```

```python
print(my_set)
```

```python
##If you dont  want an error means if present remove else no error for that we use *discard* method

my_set.discard(11)
print(my_set)
```

```python
##pop method
##Set follows first in first out principle 
removed_element = my_set.pop()
print(removed_element)
print(my_set)
```

```python
##Clear all the elements

my_set.clear()
print(my_set)
```

```python
##Set Membership Test

my_set = {1,2,3,4,5}
print(3 in my_set)
print(10 in my_set)
```

```python
##Mathematical Operations
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

###Union
union_set = set1.union(set2)
print(union_set)


###Intersection
intersection_set = set1.intersection(set2)
print(intersection_set)


###perform that operation and update that set 
set1.intersection_update(set2)
print(set1)




```

```python
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

##Difference 
print(set1.difference(set2))
print(set2.difference(set1))

###
set1
```

```python
##Symmetrice Difference

print(set1.symmetric_difference(set2))
```

```python
##Sets Methods

set1 = {1,2,3}
set2 = {3,4,5}

##is subset
print(set1.issubset(set2))

##is superset

print(set1.issuperset(set2))  ## ckeck if all elements of set2 are present in set1
```

```python
###Counting unique words and text

text = "In this tutorial we are discussing about sets . this is nice"   ##Python is case sensitive (This and this are different)
words = text.split()

##Convert list of words to set to get unique words

unique_words = set(words)
print(unique_words)
print(len(unique_words))
```
