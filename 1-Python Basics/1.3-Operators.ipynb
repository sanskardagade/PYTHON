**Logical operators**


```python
##And , Not, Or


#AND -> Both true then only true

X = True
Y = True
result = X and Y
print(result)


```

```python
#Or -> Both false then only false

X = False
Y = False
result = X or Y
print(result)

```

```python
#Not operator
X = True
not X
```
