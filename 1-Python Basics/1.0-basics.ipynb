### Syntax refers to the set of rules that defines the combination of symbols that are considered to be correctly structured programs in a language. In simpler terms , syntax is about the correct arrangement of words 

Semantics refers to the meaning ot interpretataion of the symbols, characters and commands in a language. It is about what the code is supposed to do when it runs.

```python
##Baic syntax Rules in python 
##case sensitivity - Python is case sensitive. For example, the variables 'myVar' and 'myvar' would be considered different.
##Indentation - Python uses indentation to define blocks of code. Consistent indentation is crucial for defining the structure of the code.
##Comments - Comments in Python start with the '#' symbol and extend to the end of the line.
##Line continuation - Python allows for line continuation using a backslash (\) or by enclosing expressions in parentheses, brackets, or braces.


name = "sanskar"
Name = "dagade"

print(name)  # Output: sanskar
print(Name)  # Output: dagade
```

Indendation in python is used to define the structure and hierachy of the code. Unlike many other programming languages that use braces{} to delinmit blocks of code, Python uses indentation to determine the grouping of statements. This means the statements within a block must be indented at the same level. 


```python
##Indentation 
##Python uses indentation to define blocks of code. Consistent use of spaces (commonly 4 spaces) is recommended.

age = 32 
if age>30:
    print(age + 5)

print(age)
```

```python
##This is single line comment
print("Hello, World!")  # This is an inline comment
```

```python
##line continuation 
##use a backslash (\) to continue a statement on the next line.
total = 1 + 2 + 3 + \
        4 + 5 + 6 + \
        7 + 8 + 9

print(total)
```

```python
##Multiple statements on  a single line

x=5; y=10; z=x+y
print(z)  # Output: 15

```

```python
##Understanding Semantics in Python
##variable assignmtent - In Python, variables are assigned using the '=' operator. The variable name is on the left side, and the value to be assigned is on the right side.
##Data types - Python has several built-in data types, including integers, floats, strings,
##and booleans. The type of a variable is determined at runtime based on the value assigned to it.

age = 32 ##age is an integer 
name = "sanskar" ##name is a string


```

```python
##Typeinference
type(age)  # Output: <class 'int'>
```

```python
type(name)  # Output: <class 'str'>
```

```python
##Type inference
print(type(age))  # Output: <class 'int'>
```

```python
variable = 10
print(type(variable))  # Output: <class 'int'>
variable = "Hello"
print(type(variable))  # Output: <class 'str'>
```
