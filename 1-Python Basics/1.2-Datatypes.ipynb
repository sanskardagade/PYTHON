**Datatypes**:
    Datatypes are the classification of data which tells teh compiler or interpreter how the programmer intends to use the data.
    They determine the type of operation that can be performed on the data the values the data can take and the memory needed to store the data 
    

Importance of data types in programming
1) Datatypes ensure that data is stored in efficient way
2) They help in performing correct operations on data
3) Proper use of data types can prevent errors and bugs in program
