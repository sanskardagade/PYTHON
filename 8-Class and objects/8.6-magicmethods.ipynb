## Magic Methods

    Magic methods in python also known as dunder methods(double underscore method), are specially methods that start and end with double underscores. These methods enable you to define behaviour of objects for built-in operations, such as arithmetic operations, comparisons and more.

    

##### Magic Methods are predifined methods in python that you can override to change the behaviour of your objects. Some common magic methods include:
        

```python
'''
__init__ : initializes a new instance of object 
__str__ : returns a string representation of an object
__repr__ : returns an official string representation of an object 
__len__ : returns the length of an object 
__getitem__ : Gets an item from the container 
__setitem__ : sets an item in the contatier


'''
```

```python
class Person:
    pass

person = Person()
dir(person)   ##This all are called magic methods 
```

```python
##Basic methods

class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age 

person = Person("sanskar",22)
print(person)
        
```

```python
##Basic methods

class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age 

        def __str__(self):
            return f"{self.name},{self.age} years old"

person = Person("sanskar",22)
print(person)
        
```

```python
##Basic methods

class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age 

    def __str__(self):
        return f"{self.name},{self.age} years old"

    def __repr__(self):
        return f"Person('name = {self.name}',age = {self.age})"

s = Person("sanskar",22)
print(s)
print(repr(s))
        
```
