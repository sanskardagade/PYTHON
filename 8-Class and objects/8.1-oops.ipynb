```python
##Instance variables and methods
##Instance variables are their properties means name 
##and methods are their working propertes means dog barks 

##Initializing the class with constructor 

class Dog:
    ##constructor 
    def __init__(self,name,age): ##name and age are parameter ##self is a keyword you can use your name also
        self.name = name     ##self.name this name is the instance variable of dog class
                        ##and the next name is the parameter while passing the object
        self.age = age 

##create objects 
dog1 = Dog('kutta',3)
print(dog1)    ##We will be able to create dog object at specific memory location
print(dog1.name)
print(dog1.age)            
        
```

```python
dog2 = Dog('janwar',4)   
print(dog2)
print(dog2.name)                         ##we are able to access the instance variable
print(dog2.age)
```

```python
##define a class with instance methods

class Dog:
    def __init__(self,name,age):
        self.name = name 
        self.age = age


    def bark(self):
        print(f"{self.name} says woof")

dog1 = Dog("kutta",3)
dog1.bark()
dog2 = Dog("kuttoooo",8)
dog2.bark()
```

```python
##Modelling a bank account

##Define a class for bank account

class BankAccount:
    def __init__(self,owner,balance=0):  ##owner and balance are attribute
        self.owner = owner
        self.balance = balance

    def deposit(self,amount):
        self.balance += amount
        print(f"{amount} is deposited. New balance is {self.balance}")

    def withdraw(self,amount):
        if amount>self.balance:
            print("Insufficeint funds")

        else:
            self.balance -= amount
            print(f"{amount} is withdrawn. New balance is {self.balance}")
        
    def get_balance(self):
        return self.balance
    





##Create an object
account = BankAccount("sanskar",5000)
print(account.balance)

##call instance method 
account.deposit(100)

account.withdraw(300)

print(account.get_balance())

        
```
