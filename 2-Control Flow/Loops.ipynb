```python
##while loop

##The while loop continues to excute as long as the condition is true 


count = 0
while count<5:
    print(count)
    count = count + 1
```

```python
count = 0
while count % 2 == 0:
    print(count)
    count = count + 1
```

```python
##Loop control statements

##break
##The break statement exists the loop prematurely

##break statement
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
##continue
##The continue statement skips the current iteration and continues with the next

for i in range(10):
    if i%2 == 0:
        continue
    print(i)
```

```python
##pass
##The pass statement is null operator and it does nothing

for i in range(5):
    if i==3:
        print("The number is ",i)
        pass
    print(i)
```

```python
##Nested loops
## a loop inside another loop is called nested loop
##f is formatting option 

for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

```python
##Examples - Calculate the sum of first N natural numbers using a while and a for loop 

##while loop 

n = 10
sums = 0
count = 1

while count <= n:
    sums = sums + count
    count = count + 1

print("Sum of first", n, "natural numbers using a while loop is:", sums)
```

```python
##for loop
n = 10
sums = 0
for i in range(11):
    sums = sums + i

print(sums)
```

```python
##Example : Prime number between 1 to 100

for nums in range(1, 101):
    if nums > 1:
        for i in range(2, nums):
            if nums % i == 0:
                break
        else:
            # This else belongs to the 'for', not the 'if'
            print(nums)
```

Conclusion:
    Loops are powerful constructs in Python that allow you to excute a block of code multiple times. By understanding and using for and while loops, along with loop control statements like break,continue and pass you can handle a wide range of programming tasks efficiently.
