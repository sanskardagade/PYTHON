```python
## if statement
age = 20

if age>18:
    print("you are allowed to vote in election")
```

```python
##else
##The else statement excutes a block of code if the condition in the if statement is false

age = 16

if age>18:
    print("you are allowed to vote in election")
else:
    print("you are a minor")
```

```python
## elif
## The elif statement allows you to check multiple conditions. It stands for else if 

age = 20

if age<13:
    print("you are a child")
elif age<18:
    print("You are a teenager")
else:
    print("You are an adult") 
```

```python
##Nested Conditional Statements

#You can place one or more if,elif,else statements inside another if,elof,else statements

##number even odd or negative


num = int(input("Enter a number"))

if num>0:
    print("The number is positive")

    if(num%2==0):
        print("The number is even")
    else:
        print("The number is odd")

else:
    print("The number is zero or negative")
```

```python
##Practical Examples

##Determine if a year is leap year using nested conditional statement

year = int(input("Enter the year"))

if year%4==0:
    if year%100==0:
        if year%400==0:
            print(year,"is a leap year")
        else:
            print(year,"is not a leap year")
    else:
        print(year,"is a leap year")
else:
    print(year,"is not a leap year")
```

```python
##Assignment
##Simple Calculator

num1 = float(input("Enter first number"))
num2 = float(input("Enter second number"))
operation = input("Enter operation(+,-,*,/):")

if operation == '+':
    result = num1 + num2

elif operation == '-':
    result = num1 - num2

elif operation == '*':
    result = num1 * num2

elif operation == '/':
    if num2!=0:
        result = num1/num2
    else:
        result = "Error! Division by zero"

else:
    result = "Invalid operation"

print("Result : ",result)
```
