```python
##why functions?
num = 24
if num % 2 == 0:
    print("the number is even")
else:
    print("the number is odd")
    
```

```python
def even_or_odd(num):
    """This function finds even or odd"""
    if num % 2 == 0:
        print("The number is even")
    else:
        print("The number is odd")
    
```

```python
##Call the function 

even_or_odd(23)
```

```python
##Function with multiple parameter
def add(a,b):
    return a + b


result = add(2,3)
print(result)
```

```python
##Default parameters

def greet(name):
    print(f"Hello {name}")


greet("sanskar")
```

```python
##Default parameters

def greet(name="Guest"):
    print(f"Hello {name}")


greet()
greet("sanskar")
```

```python
##Variable Length Arguments
##Positional and keyword arguments


##Positional agruments

def print_numbers(*args):                                ##instead of args any name can be there acccording to us
    for number in args:
        print(number)                                    ##This is called positional arguments
```

```python
print_numbers(1,2,3,4,54,5,6,6,7,6,6,7,"snd")
```

```python
##keywords arguments
##all the parameters would be in key value pairs  we can access it like dictionaries 

def print_details(**kwargs): 
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(name="sanskar",age=22)
```

```python
def combined(*args,**kwargs):
    for val in args:
        print(f"Positional argument : {val}")

    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
combined(1,2,3,4,4,5,6,7,8,name="sanskar",age=22)  ##You cannot write positional arguments after keyword arguments
```
