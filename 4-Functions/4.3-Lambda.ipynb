**Lambda Functions in python**

Lambda functions are small anonymous(A function without a name) fucntions defined using lambda keyword. They can have any number of arguments but only one expression. They are commonly used for short operations or as arguments to higher order functions.


Only single expression can be converted to lambda function

```python
##Syntax
# lambda arguments : expression
```

```python
def addition(a,b):
    return a+b
```

```python
addition(2,3)
```

```python
add = lambda a,b:a+b
print(type(add))
print(add(2,3))
```

```python
def even(num):
    if num%2 ==0 :
        return True
even(24)
```

```python
even1 = lambda num : num % 2 == 0
print(type(even1))
print(even1(23))
```

```python
def additions(x,y,z):
    return x + y + z

additions(2,3,4)
```

```python
adds = lambda x,y,z : x + y + z

print(adds(2,3,4))
```

```python
##map() - applies a function to all items in a list 

numbers=[1,2,3,4,5,6]
def square(number):
    return number**2

square(4)
```

```python
list(map(lambda x:x**2,numbers) )   ##It uses heavy loading means it has not completely initalized 
```
