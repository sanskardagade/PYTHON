##  Reading data from Differenet Sources

```python
import pandas as pd
from io import StringIO

json_data = '''
[
  {"employee_name": "James", "email": "james@gmail.com", "job_profile": "Team Lead"},
  {"employee_name": "Michael", "email": "michael@gmail.com", "job_profile": "Senior Developer"}
]
'''
df = pd.read_json(StringIO(json_data))
print(df)
```

```python
df.to_json()
```

```python
df.to_json(orient='index')
```

```python
df.to_json(orient='records')
```

```python
import pandas as pd
import requests

url = "https://en.wikipedia.org/wiki/Mobile_country_code"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

html = requests.get(url, headers=headers).text

tables = pd.read_html(html, match="Country")
df = tables[0]

print(df.head())
```
