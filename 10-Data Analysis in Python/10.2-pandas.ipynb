## Pandas - Data frames and series 

Pandas is powerful data manipulation library in python, widely used for data analysis and cleaning. It provides two primary data structure: Series and Data Frame.

A Pandas serues us a 1D array like objects that can hold any data type. It is similar to column in table. The basic difference between a series and dataframe is that series is 1D while dataframe is 2D mutable and can contain heterogeneous data .

```python
import pandas as pd
```

```python
##Series
##A pandas series is a one dimesnional array like object that can hold any data type. It is similar to colume in table..


import pandas as pd

data = [1,2,3,4,5]
series = pd.Series(data)
print("Series: \n",series)
print(type(series))
```

```python
## Create a series from dictionary 

data = {'a':1,'b':2,'c':3}
series_dict = pd.Series(data)
print(series_dict)
```

```python
data = [10,20,30]
index = ['a','b','c']
pd.Series(data,index = index)
```

```python
##Dataframe 
##Create a dataframe from a dictionary of list 

data = {
    'Name':["sanskar","kruish","sans"],
    'Age': [35,40,40],
    'City' : ['Kopargaon','Lasur','Aurangabad']
}

df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
import numpy as np 
np.array(df)                        ##we can do it like this also 
```
