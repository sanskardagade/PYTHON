## Reference counting

Reference counting is the primary method Python uses to manage memory. Each object in python maintains a count of references pointing to it. When the reference count drops to zero, the memory occupied by the object is deallocated. 

```python
import sys 

a = []

print(sys.getrefcount(a))       ## 2 (one reference is from 'a' and one from 'getrefcount()')
```

```python
b = a 
print(sys.getrefcount(b))
```

```python
del b 
print(sys.getrefcount(b))
```

```python
print(sys.getrefcount(a))
```

#### Garbage collection


Python includes a cyclic garbage collector to handle reference cycles. Reference cycles occur when objects reference each other, preventing their reference counts from reaching zero.



```python
import gc
##enable garbage collection 
gc.enable()
```

```python
gc.disable()
```

```python
gc.collect()
```

```python
### Get garbage collection stats
print(gc.get_stats())
```

```python
## get uncreachable objects

print(gc.garbage)
```

## Memory Management Best Practices

1. Use Local Variables: Local variables have a shorter lifespan and are freed sooner than global variables. 
2. Avoid Circular References: Circular references can lead to memory leaks if not properly managed. 
3. Use Generators: Generators produce items one at a time and only keep one item in memory at a time, making them memory efficient. 
4. Explicitly Delete Objects: Use the del statements to delete variables and objects explicetly. 
5. Profile Memory Usuage: Use memory profiling tools like tracemalloc and memory_profiler to identify memory leaks and optimize memory usuage.

```python
## Handled Circular Reference

import gc

class MyObject:
    def __init__(self,name):
        self.name = name
        print(f"Object {self.name} created")
    
    def __del__(self):
        print(f"Object {self.name} deleted")


##Create self reference 

obj1 = MyObject("obj1")
obj2 = MyObject("obj2")


obj1.ref = obj2
obj2.ref = obj1


del obj1
del obj2



##Manually trigger the garbage collection 

gc.collect()

```

```python
##Generators for memory efficiency 

## Generators allows you to produce items one at a time, using memory efficiently by only keeping one item in memory at a time.

def generate_numbers(n):
    for i in range (n):
        yield i

## using the generator 

for num in generate_numbers(100000):
    print(num)
    if(num>10):
        break
```

```python
## Profiling Memory usuage with tracemalloc

import tracemalloc

def create_list():
    return [i for i in range(10000)]

def main():
    tracemalloc.start()

    create_list()

    snapshot = tracemalloc.take_snapshot()
    top_stats = snapshot.statistics('lineno')

    print('[Top 10]')
    for stat in top_stats[:10]:
        print(stat)

```

```python
main()
```
